```python
def dfs(graph, v, visited):
    #현재 반문한 노드 방문으로 표시
    visited[v] = True
    print(v, end = ' ')
    
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited) 
                     
visited = [False] * 9

graph = [[],
         [2,3,8],
         [1,7],
         [1,4],
         [3,5],
         [3,4],
         [7],
         [2,6,8],
         [1,7]
]

dfs(graph, 1, visited)

```
