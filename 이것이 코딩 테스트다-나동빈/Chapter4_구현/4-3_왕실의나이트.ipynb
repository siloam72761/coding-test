```python
N = input()
row = int(N[1])
column = int(ord(N[0])) - int(ord('a')) + 1

knight_move = [(-2,-1), (-1,-2), (1,-2), (2,-1), (2,1), (1,2), (-1,2), (-2,1)]
count = 0

for move in knight_move:
    next_row = row + move[0]
    next_column = column + move[1]
    if next_row >=1 and next_row <=8 and next_column >=1 and next_column <=8:
            count += 1

print(count)
```
