```python
N, M = map(int, input().split()) # N * M 맵 생성
x, y, dir = map(int, input().split()) # (현재 좌표 + 방향)
topography = [list(map(int, input().split())) for _ in range(N)]

d = [[0] * M for _ in range(N)]
d[x][y] = 1 # 현재 좌표


#방향
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

#왼쪽으로 방향 전환
def turn_left():
    global dir
    dir -= 1
    if dir == -1:
        dir = 3


count = 1
turn_time = 0


while(True):
    turn_left()
    nx = x + dx[dir]
    ny = y + dy[dir]

    #왼쪽으로 전진 가능한 경우
    if d[nx][ny] == 0 and topography[nx][ny] == 0:
        d[nx][ny] = 1
        x = nx
        y = ny
        count += 1
        turn_time = 0
        continue
    #다른 이유로 막혀 있는 경우
    else:
        turn_time += 1
    #4번 다 막혔을 경우
    if turn_time == 4:
        nx = x - dx[dir]
        ny = y - dy[dir]
        #뒤에 땅이 있으면 이동
        if topography[nx][ny] == 0:
            x = nx
            y = ny
            turn_time = 0
        #뒤가 바다로 막혀있을 경우
        else:
            break
        
print(count)
```
