```python
def binary_search(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2

        if array[mid] == target:
            return mid
        #중간 값보다, 타겟이 더 낮은 경우 왼쪾 확인
        elif array[mid] > target:
            end = mid - 1
        else:
            start = mid + 1

    return None
            

#N개의 부품
N = int(input())
arr = list(map(int, input().split()))
arr.sort()

M = int(input())
items = list(map(int,input().split()))

for i in items:
    result = binary_search(arr, i ,0, N-1)
    if result != None:
        print('yes', end = ' ')
    else:
        print('no', end = ' ')
```
